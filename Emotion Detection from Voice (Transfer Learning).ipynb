```python
image_count = len(list(FOLDER.glob('*/*.png')))
print(image_count)
```

```python
batch_size = 32
img_height = 217
img_width = 334
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    FOLDER,
    validation_split=0.25,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)
```

```python
validation_ds = tf.keras.utils.image_dataset_from_directory(
    FOLDER,
    validation_split=0.25,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
for image_batch, labels_batch in train_ds:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)
```

```python
normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
# Notice the pixel values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
normalized_validation_ds = validation_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch_v, labels_batch_v = next(iter(normalized_validation_ds))
first_image = image_batch_v[0]
# Notice the pixel values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
validation_ds = validation_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
resnet_model = keras.Sequential()

pretrained_model = tf.keras.applications.ResNet50(include_top=False, 
                                                 input_shape=(217, 334, 3),
                                                  classes=4, 
                                                  pooling='max', 
                                                  weights='imagenet')
for layer in pretrained_model.layers:
    layer.trainable = False
    
resnet_model.add(pretrained_model)
resnet_model.add(layers.Flatten())
resnet_model.add(layers.Dense(128, activation="relu"))
resnet_model.add(layers.Dense(4, activation="softmax"))
```

```python
resnet_model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), 
                    loss='sparse_categorical_crossentropy', 
                    metrics=['accuracy'])
```

```python
history = resnet_model.fit(train_ds, epochs=100)
```

```python
resnet_model.evaluate(validation_ds)
```

### Feature Extractor EfficientNet

```python
efficient_model = keras.Sequential()

efficient_base = tf.keras.applications.efficientnet_v2.EfficientNetV2L(include_top=False, 
                                                  input_shape=(217, 334, 3),
                                                  classes=4, 
                                                  pooling='max', 
                                                  weights='imagenet')
```

```python
efficient_base.layers[1028]

for layer in efficient_base.layers:
    if layer == "keras.layers.pooling.global_max_pooling2d.GlobalMaxPooling2D object at 0x000001FC67170C88":
        layer.trainable = False
```

```python
def extract_features(datagen, sample_count):
    
    features = np.zeros(shape=(sample_count, 1280))
    labels = np.zeros(shape=(sample_count))
    
    i=0
    for inputs_batch, labels_batch in datagen:
        features_batch = efficient_base.predict(inputs_batch)
        features[i * batch_size : (i + 1) * batch_size] = features_batch
        labels[i * batch_size : (i + 1) * batch_size] = labels_batch
        i += 1
        if i * batch_size >= sample_count:
            break
    return features, labels

train_features, train_labels = extract_features(normalized_ds, 6655)
test_features, test_labels = extract_features(normalized_validation_ds, 2218)
```

```python
np.savetxt("E:/train_features.csv", train_features, delimiter=",")
```

```python
import csv
filename = "test_features_flat.csv"
f = open(filename, "w")
wr = csv.writer(f, delimiter=",")
for i in range(len(test_features)):
    data = test_features[i]
    labels = test_labels[i]
    wr.writerow((data, labels))
f.close()
```
